```python
def HaarExamples(vidFile,delay=220):
    #OpenCV's pre-trained Classifiers
    #Face == Blue
    Face_cascade = cv2.CascadeClassifier(os.getcwd()+'\\haarcascade_frontalface_default.xml')
    #Smile == Green
    Smile_cascade = cv2.CascadeClassifier(os.getcwd()+'\\haarcascade_smile.xml')
    fourcc = cv2.cv.CV_FOURCC('m', 'p', '4', 'v')
    video = cv2.VideoWriter('HaarCascadeExample.avi',fourcc,1,(1280,720))
    cap = cv2.VideoCapture(vidFile)
    while(True):
        # Capture frame-by-frame
        for i in range(90):
            ret, frame = cap.read()
            if not ret:
                cap.release()
                cv2.destroyAllWindows()
                return
                
        smallframe = cv2.resize( cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY), (0,0), fx=0.25, fy=0.25)  
        faces  =  Face_cascade.detectMultiScale(smallframe, 1.3, 5)
        smiles =  Smile_cascade.detectMultiScale(smallframe, 1.3, 5)
        for (x,y,w,h) in faces:
            cv2.rectangle(frame,(4*x,4*y),(4*(x+w),4*(y+h)),(255,0,0),5)
        for (x,y,w,h) in smiles:
            cv2.rectangle(frame,(4*x,4*y),(4*(x+w),4*(y+h)),(0,255,0),5)
        if not ret:
            break
        # Display the resulting frame
        cv2.imshow('Haar Cascade Example',frame)
        video.write(frame)
        if cv2.waitKey(delay) & 0xFF == ord('q'):
            break
    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()
    video.release()
HaarExamples(os.getcwd()+'\\Source Media\\BabySmile.mp4')
```

```python
#Feature Point, Harris, Example Video, zoomed in to see more features
def FeaturePointsHarris(vidFile):
    cap = cv2.VideoCapture(vidFile)
    fourcc = cv2.cv.CV_FOURCC('m', 'p', '4', 'v')
    video = cv2.VideoWriter('Harris Corner Feature Detector.avi',fourcc,1,(450,280))
    while(True):
        #Show example of a frame from video from every ~ three seconds
        for i in range(3):
            ret, frame = cap.read()
            if not ret:
                cap.release()
                cv2.destroyAllWindows()
                return 225
        #Harris Corner Detection
        frame_harris = frame[440:840,135:585]
        gray = cv2.cvtColor(cv2.medianBlur(frame[440:840,135:585],5),cv2.COLOR_BGR2GRAY)
        gray_f = np.float32(gray)
        dst_harris = cv2.cornerHarris(gray_f,2,3,0.04)

        # Threshold for an optimal value, it may vary depending on the image.
        frame_harris[dst_harris>0.01*dst_harris.max()]=[0,0,255]
        cv2.imshow('Harris Corner',frame_harris)
        video.write(frame_harris)
        if cv2.waitKey(1000) & 0xFF == ord('q'):
            break
    # When everything done, release the capture
    cap.release()   
    cv2.destroyAllWindows() 
    video.release()
     
FeaturePointsHarris(os.getcwd()+'\\Source Media\\BabySmile.mp4')
```

```python
#Feature Point Shi-Tomasi Corner Detector Example Video, zoomed in to see more features
def FeaturePointsShi_Tomasi(vidFile):
    cap = cv2.VideoCapture(vidFile)
    fourcc = cv2.cv.CV_FOURCC('m', 'p', '4', 'v')
    video = cv2.VideoWriter('Shi-Tomasi Corner Detector Example Video.avi',fourcc,1,(450,280))
    while(True):
        #Show example of a frame from video from every ~ three seconds
        for i in range(3):
            ret, frame = cap.read()
            if not ret:
                cap.release()
                cv2.destroyAllWindows()
                return 
        #Harris Corner Detection
        frame_shi = frame[440:840,135:585]
        gray = cv2.cvtColor(frame_shi,cv2.COLOR_BGR2GRAY)
        corners = cv2.goodFeaturesToTrack(gray,70,0.05,10)
        corners = np.int0(corners)

        for i in corners:
            x,y = i.ravel()
            cv2.circle(frame_shi,(x,y),3,255,-1)


        cv2.imshow('Shi-Tomasi Corner Detector',frame_shi)
        video.write(frame_shi)
        if cv2.waitKey(20) & 0xFF == ord('q'):
            break
    cv2.destroyAllWindows()  
    cap.release()    
    video.release()       
FeaturePointsShi_Tomasi(os.getcwd()+'\\Source Media\\BabySmile.mp4')
```

### SURF Detector (Speeded Up Robust Feature)

```python
#Feature Point SURF Detector Example Video, zoomed in to see more features
def FeaturePointsSURF(vidFile,delay=1000):
    cap = cv2.VideoCapture(vidFile)
    surf = cv2.SURF()
    fourcc = cv2.cv.CV_FOURCC('m', 'p', '4', 'v')
    video = cv2.VideoWriter('SURF Detector (Speeded Up Robust Feature).avi',fourcc,1,(450,280))
    while(True):
        #Show example of a frame from video from every ~ three seconds
        for i in range(3):
            ret, frame = cap.read()
            if not ret:
                cap.release()
                cv2.destroyAllWindows()
                return 
        #SURF
        frame_SURF = cv2.medianBlur(frame[440:840,135:585],5)
        gray = cv2.cvtColor(frame_SURF,cv2.COLOR_BGR2GRAY)
        # SURF extraction of key points
        keypoints = surf.detect(gray,None)

        #Draw a small red circle with the desired radius
        #at the (x, y) location for each feature found
        for kp in keypoints:
            x = int(kp.pt[0])
            y = int(kp.pt[1])
            cv2.circle(frame_SURF, (x, y), 2, (0, 0, 255))
        cv2.imshow('SURF',frame_SURF)
        video.write(frame_SURF)
        if cv2.waitKey(delay) & 0xFF == ord('q'):
            break
    cv2.destroyAllWindows()  
    cap.release()    
    video.release()   
FeaturePointsSURF(os.getcwd()+'\\Source Media\\BabySmile.mp4',delay=10) #Show speed of SURF algorithim
```

### Comparison of Harris, Shi, and SURF Algorithims Speed

```python
def FeatureDescriptorAnalytics(vidFile,delay=10,repeat=1):
    from timeit import Timer
    timeDict = {'Harris':[],
                'Shi':[],
                'SURF':[],
                'SIFT':[]}
    surf = cv2.SURF()
    sift = cv2.SIFT()
    cap = cv2.VideoCapture(vidFile)
    
    while(True):
        ret, frame = cap.read()
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        #Now Time each feature descript to get timing per each frame for analysis
        Harris = Timer(lambda: cv2.cornerHarris(gray,2,3,0.04))
        Shi = Timer(lambda: cv2.goodFeaturesToTrack(gray,70,0.05,10))
        SURF = Timer(lambda: surf.detect(gray,None))
        SIFT = Timer(lambda: sift.detect(gray,None))
        #Append the time per function call on present frame
        timeDict['Harris'].append(Harris.timeit(repeat)) 
        timeDict['Shi'].append(Shi.timeit(repeat))
        timeDict['SURF'].append(SURF.timeit(repeat))
        timeDict['SIFT'].append(SIFT.timeit(repeat))
        cv2.imshow('Video Frame',frame)
        if cv2.waitKey(delay) & 0xFF == ord('q'):
            break
    cv2.destroyAllWindows()  
    cap.release()  
    return timeDict

timeDict = FeatureDescriptorAnalytics(os.getcwd()+'\\Source Media\\BabySmile.mp4')
for k in timeDict.keys():
    plt.plot(range(len(timeDict[k])),timeDict[k],label=k)
plt.legend()

```

### Video Stabalization using frame to frame movement, No smoothing of movement

```python
def StabalizationControl(vidFile,delay=10,skip = 0, MIN_MATCH_COUNT=50):

    """ Matches the SURF keypoints of refernce image to frames in video to
        produce a stabalized video
        Frame to Frame calculation of transformation Matrix
    """
    cap = cv2.VideoCapture(vidFile)
    # Create SURF object. You can specify params here or later.
    # Here I set Hessian Threshold to 400
    surf = cv2.SURF(1000)
    kp2, des2 = None, None
    FLANN_INDEX_KDTREE = 0
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = dict(checks = 50)
    i=0
    while(True):
        ret, frame = cap.read()
        if skip>0:
            for _ in range(skip):
                ret, frame = cap.read()
        
        if not ret:
            cap.release()
            cv2.destroyAllWindows()
            return 
        gray = cv2.cvtColor(cv2.medianBlur(frame,5),cv2.COLOR_BGR2GRAY)
        kp1, des1  = surf.detectAndCompute(frame,None)
        if  kp2:   #Warp perspective via homogprahy
            #Flann 
            #flann = cv2.FlannBasedMatcher(index_params, search_params)
            #matches = flann.knnMatch(des1,des2,k=2)
            
            # BFMatcher with default params
            bf = cv2.BFMatcher()
            matches = bf.knnMatch(des1,des2, k=2)
            # store all the good matches as per Lowe's ratio test.
            good = []
            for m,n in matches:
                if m.distance < 0.7*n.distance:
                    good.append(m)
            if len(good)>MIN_MATCH_COUNT:
                src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
                dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)

                M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)
                h,w,c = frame.shape
                frame = cv2.warpPerspective(frame, M, (w,h))
            else:
                print "Not enough matches are found - %d/%d" % (len(good),MIN_MATCH_COUNT)
        frame = cv2.resize( frame, (0,0), fx=0.5, fy=0.5)  
        cv2.imshow('Stabilization Control',frame)
        if cv2.waitKey(delay) & 0xFF == ord('q'):
            break       
        kp2, des2 = kp1, des1
        i+=1
        
        
    # When everything done, release the capture
    cap.release()
    cv2.destroyAllWindows()       
StabalizationControl(os.getcwd()+'\\Source Media\\BabySmile.mp4',skip=10)
```

### Now we will filter out baby via hue channel (thresholding, erosion, dilation, numpy indexing)

```python
def findBabyRevA(frame):
    #Hue < 30 is probably a baby
    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    h,s,v = cv2.split(frame)
    ret,thresh = cv2.threshold(h,32,255,cv2.THRESH_BINARY_INV)
    cv2.imshow('Baby via Hue',thresh)
 
def findBabyRevB(frame):
    #Hue < 30 is probably a baby
    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    h,s,v = cv2.split(frame)
    ret,thresh = cv2.threshold(h,32,255,cv2.THRESH_BINARY_INV)
    kernel = np.ones((4,4),np.uint8)
    thresh = cv2.erode(thresh,kernel,iterations = 2)
    contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    baby_i = np.argmax([cv2.contourArea(cnt) for cnt in contours])
    baby_mask = np.zeros_like(thresh)
    cv2.drawContours(baby_mask,[contours[baby_i]],0,255,-1)
    kernel = np.ones((3,3),np.uint8)
    baby_mask = cv2.dilate(baby_mask,kernel,iterations = 2)
    cv2.imshow('Baby via Hue',baby_mask)

def findBabyRevC(frame):
    #Hue < 30 is probably a baby
    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    h,s,v = cv2.split(frame)
    ret,thresh = cv2.threshold(h,32,255,cv2.THRESH_BINARY)
    kernel = np.ones((4,4),np.uint8)
    thresh = cv2.erode(thresh,kernel,iterations = 2)
    contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    baby_i = np.argmax([cv2.contourArea(cnt) for cnt in contours])
    baby_mask = np.zeros_like(thresh)
    cv2.drawContours(baby_mask,[contours[baby_i]],0,255,-1)
    kernel = np.ones((3,3),np.uint8)
    baby_mask = cv2.dilate(baby_mask,kernel,iterations = 2)
    return baby_mask
    
        
cap = cv2.VideoCapture(os.getcwd()+'\\Source Media\\BabySmile.mp4')
while(True):
    ret, frame = cap.read()
    frame =  cv2.resize( frame, (0,0), fx=0.3, fy=0.3)  
    findBabyRevB(frame)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break   
cap.release()
cv2.destroyAllWindows() 
```
